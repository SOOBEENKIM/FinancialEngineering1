Problem2_Q1

```python
def minus(a,b):
    return a-b
data_1 = data[['SAMSUNG ELECTRONICS', 'NAVER','KOSPI', 'KOR10Y']]
data_1['Samsung-KOR10Y'] = data_1.apply(lambda x: minus(x['SAMSUNG ELECTRONICS'], x['KOR10Y']), axis=1)
data_1['KOSPI-KOR10Y'] = data_1.apply(lambda x: minus(x['KOSPI'], x['KOR10Y']), axis=1)
data_1['Naver-KOR10Y'] = data_1.apply(lambda x: minus(x['NAVER'], x['KOR10Y']), axis=1)
mean_1 = np.mean(data_1)
data_1.head()
cov = data_1.cov()
```

```python
# cov에서 열 이름이 제대로 매핑되는지 확인
print("Samsung-KOR10Y var: {:6f}, Naver-KOR10Y var : {:6f}, KOSPI-KOR10Y var : {:6f}".format(
    cov.at['Samsung-KOR10Y', 'Samsung-KOR10Y'], 
    cov.at['Naver-KOR10Y', 'Naver-KOR10Y'], 
    cov.at['KOSPI-KOR10Y', 'KOSPI-KOR10Y']))

print("Samsung-KOR10Y cov: {:6f}, Naver-KOR10Y cov : {:6f}, KOSPI-KOR10Y cov : {:6f}".format(
    cov.at['Samsung-KOR10Y', 'KOSPI-KOR10Y'], 
    cov.at['Naver-KOR10Y', 'KOSPI-KOR10Y'], 
    cov.at['KOSPI-KOR10Y', 'KOSPI-KOR10Y']))
```

Problem2_Q1_result_summary

```python
# 평균값 및 공분산 출력
mean_1 = data_1.mean()
cov = data_1.cov()

# 평균값 출력
print(f"aver\nSamsung-KOR10Y: {mean_1['Samsung-KOR10Y']:.6f}, Naver-KOR10Y: {mean_1['Naver-KOR10Y']:.6f}, KOSPI-KOR10Y: {mean_1['KOSPI-KOR10Y']:.6f}")

# 분산 출력
print(f"var\nSamsung-KOR10Y: {cov.at['Samsung-KOR10Y', 'Samsung-KOR10Y']:.6f}, Naver-KOR10Y: {cov.at['Naver-KOR10Y', 'Naver-KOR10Y']:.6f}, KOSPI-KOR10Y: 
      {cov.at['KOSPI-KOR10Y', 'KOSPI-KOR10Y']:.6f}")

# 공분산 출력
print(f"cov\nSamsung-KOR10Y with KOSPI-KOR10Y: {cov.at['Samsung-KOR10Y', 'KOSPI-KOR10Y']:.6f}, Naver-KOR10Y with KOSPI-KOR10Y: 
      {cov.at['Naver-KOR10Y', 'KOSPI-KOR10Y']:.6f}")

# CAPM에 필요한 beta (b)와 alpha (a) 계산
b_samsung = cov.at['Samsung-KOR10Y', 'KOSPI-KOR10Y'] / cov.at['KOSPI-KOR10Y', 'KOSPI-KOR10Y']
b_naver = cov.at['Naver-KOR10Y', 'KOSPI-KOR10Y'] / cov.at['KOSPI-KOR10Y', 'KOSPI-KOR10Y']

a_samsung = mean_1['Samsung-KOR10Y'] - b_samsung * mean_1['KOSPI-KOR10Y']
a_naver = mean_1['Naver-KOR10Y'] - b_naver * mean_1['KOSPI-KOR10Y']

# 결과 출력
print(f"b (beta)\nSamsung-KOR10Y: {b_samsung:.6f}, Naver-KOR10Y: {b_naver:.6f}")
print(f"a (alpha)\nSamsung-KOR10Y: {a_samsung:.6f}, Naver-KOR10Y: {a_naver:.6f}")

# 잔차 분산 (error variance) 계산
e_var_samsung = cov.at['Samsung-KOR10Y', 'Samsung-KOR10Y'] - b_samsung * cov.at['Samsung-KOR10Y', 'KOSPI-KOR10Y']
e_var_naver = cov.at['Naver-KOR10Y', 'Naver-KOR10Y'] - b_naver * cov.at['Naver-KOR10Y', 'KOSPI-KOR10Y']

# 잔차 분산 출력
print(f"e-var\nSamsung-KOR10Y: {e_var_samsung:.6f}, Naver-KOR10Y: {e_var_naver:.6f}")
```

```python
summary_1_df = data_1[['Samsung-KOR10Y', 'Naver-KOR10Y', 'KOSPI-KOR10Y']]
samsung_b = cov.iloc[4,5] / cov.iloc[5,5]
samsung_a = mean_1[4] - samsung_b * mean_1[5]
samsung_error_variance = cov.iloc[4,4] - (samsung_b**2) * cov.iloc[5,5]
naver_b = cov.iloc[6,5] / cov.iloc[5, 5]
naver_a = mean_1[6] - naver_b * mean_1[5]
naver_error_variance = cov.iloc[6,6] - (naver_b**2) * cov.iloc[5,5]
summary_1_df.loc['aver'] = [mean_1[4], mean_1[6], mean_1[5]]
summary_1_df.loc['var'] = [cov.iloc[4, 4], cov.iloc[6, 6], cov.iloc[5, 5]]
summary_1_df.loc['cov'] = [cov.iloc[4, 5], cov.iloc[6, 5], cov.iloc[5, 5]]
summary_1_df.loc['b'] = [samsung_b, naver_b, 1]
summary_1_df.loc['a'] = [samsung_a, naver_a, 0]
summary_1_df.loc['e-var'] = [samsung_error_variance, naver_error_variance, np.nan]
```

```python
## Summary
print(summary_1_df)
```

```python
## Adequacy
error_samsung = data_1['Samsung-KOR10Y'] - (samsung_a + samsung_b * data_1['KOSPI-KOR10Y'])
error_naver = data_1['Naver-KOR10Y'] - (naver_a + naver_b * data_1['KOSPI-KOR10Y'])
correlation = np.corrcoef(np.array([error_samsung, error_naver]))

print("상관 계수 : ", correlation[0,1])
```

Problem2_Q2

```python
def minus(a, b):
    return a-b

# 데이터 선택
data_2 = data[['KAKAO', 'SK TELECOM', 'KOSPI', 'KOR10Y']]

# 평균 및 공분산 계산
mean_2 = data_2.mean()
cov_2 = data_2.cov()

# 베타, 알파, 잔차 분산 계산
kakao_b = cov_2.loc['KAKAO', 'KOSPI'] / cov_2.loc['KOSPI', 'KOSPI']
kakao_a = mean_2['KAKAO'] - kakao_b * mean_2['KOSPI']
kakao_error_variance = cov_2.loc['KAKAO', 'KAKAO'] - (kakao_b**2) * cov_2.loc['KOSPI', 'KOSPI']

sk_b = cov_2.loc['SK TELECOM', 'KOSPI'] / cov_2.loc['KOSPI', 'KOSPI']
sk_a = mean_2['SK TELECOM'] - sk_b * mean_2['KOSPI']
sk_error_variance = cov_2.loc['SK TELECOM', 'SK TELECOM'] - (sk_b**2) * cov_2.loc['KOSPI', 'KOSPI']

# 요약 데이터프레임 생성
summary_2_df = data[['KAKAO', 'SK TELECOM', 'KOSPI']].copy()

# 평균, 분산, 공분산, 베타, 알파 및 잔차 분산 추가
summary_2_df.loc['aver'] = [mean_2['KAKAO'], mean_2['SK TELECOM'], mean_2['KOSPI']]
summary_2_df.loc['var'] = [cov_2.loc['KAKAO', 'KAKAO'], cov_2.loc['SK TELECOM', 'SK TELECOM'], cov_2.loc['KOSPI', 'KOSPI']]
summary_2_df.loc['cov'] = [cov_2.loc['KAKAO', 'KOSPI'], cov_2.loc['SK TELECOM', 'KOSPI'], cov_2.loc['KOSPI', 'KOSPI']]
summary_2_df.loc['b'] = [kakao_b, sk_b, 1]
summary_2_df.loc['a'] = [kakao_a, sk_a, 0]
summary_2_df.loc['e-var'] = [kakao_error_variance, sk_error_variance, np.nan]

# Adequacy  
error_kakao = data_2['KAKAO'] - (kakao_a + kakao_b * data_2['KOSPI'])
error_sk = data_2['SK TELECOM'] - (sk_a + sk_b * data_2['KOSPI'])
corr_2 = np.corrcoef(np.array([error_kakao, error_sk]))
print("상관 계수 : ", corr_2[0, 1])
```
